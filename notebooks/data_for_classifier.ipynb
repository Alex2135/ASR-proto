```python
request = os.path.join(SPECT_DIR, "*.jpg")
spec_files = []
for file in glob.glob(request):
    spec_files.append(file)
#pp(spec_files)
def filename_to_class(filename: str) -> int:
    basename = os.path.basename(filename)
    return int(basename[0])

cls_to_sent = {
    1: "Запустити відео обтікання будинків",
    2: "запустити відео демонстрації вихорів",
    3: "запустити відео обтікання крила",
    4: "закрити відео"
}

classes = [filename_to_class(filename) for filename in spec_files]
sentences = [cls_to_sent[idx] for idx in classes]

df = pd.DataFrame({
    "spectro_path": spec_files,
    "basename": [os.path.basename(f) for f in spec_files],
    "class": classes,
    "sentence": sentences
})
print(f"{len(df)=}")
df
```

```python
#df_cvoice_train.append(df_append, ignore_index=True)
df_cvoice_sample = df_cvoice_train.sample(len(df)//4)
for idx, row in df_cvoice_sample.iterrows():
    df_append = {
        "spectro_path": os.path.join(SPECT_TRAIN_DIR, row["spectro_path"]),
        "sentence": row["sentence"],
        "class": 0,
        "basename":  row["spectro_path"]
    }
    df = df.append(df_append, ignore_index=True)

df.shape
```

```python
from tqdm import tqdm

def img_to_array(img_path, shape=None, fill=0):
    result = None
    img = PImage.open(img_path).convert("L")
    #img = ImageOps.grayscale(img)
    if img and shape:
        result = np.zeros(shape) + fill
        arr = np.array(np.asarray(img))
        rows = arr.shape[0]
        cols = arr.shape[1]
        result[:rows, :cols] = arr
    elif img:
        result = np.array(np.asarray(img))
    return result

idx = 1
max_dim1 = 0
max_dim2 = 0
min_dim1 = 10000
min_dim2 = 10000
df_len = len(df)

print(f"{df_len=}")
id_to_imgs = {}

for row_idx in tqdm(range(df_len)):
    path = df.iloc[row_idx]["path"]
    image_array = img_to_array(path)
    id_to_imgs[row_idx] = image_array
    dim1, dim2 = image_array.shape
    max_dim1, max_dim2 = max(max_dim1, dim1), max(max_dim2, dim2)
    min_dim1, min_dim2 = min(min_dim1, dim1), min(min_dim2, dim2)
    """
    if dim1 < 768:
        print(f"dim1 < 768 = {dim1}", path)
    if dim2 < 1070:
        print(f"\tdim2 < 1070 = {dim2}", path)
    """

#plt.figure(figsize=(12, 5))
#plt.imshow(image_array, cmap=plt.get_cmap('gray')) 
print("max dim:", (max_dim1, max_dim2))
print("min dim:", (min_dim1, min_dim2))
```
