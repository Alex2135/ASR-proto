```python
import os
import glob
from pprint import pprint as pp

import numpy as np
import pandas as pd
```

```python
PATH = r"D:\ML\Speech recognition\NLP_diploma\uk"
SPECT_DIR = os.path.join(PATH, "spectrograms_classifire")
CLIPS_DIR = os.path.join(PATH, "clips_classifire")
CSV_PATH = os.path.join(PATH, "classifier.csv")
print(PATH, SPECT_DIR, CLIPS_DIR, CSV_PATH, sep="\n")
```

```python
request = os.path.join(SPECT_DIR, "*.jpg")
spec_files = []
for file in glob.glob(request):
    spec_files.append(file)
#pp(spec_files)
def filename_to_class(filename: str) -> int:
    basename = os.path.basename(filename)
    return int(basename[0])

classes = [filename_to_class(filename) for filename in spec_files]

df = pd.DataFrame({
    "path": spec_files,
    "basename": [os.path.basename(f) for f in spec_files],
    "class": classes
})
df
```

```python
df.groupby(by="class").count()
```

```python
df.to_csv(CSV_PATH)
```
